```python
!pip install dimcli -U --quiet

import dimcli
from dimcli.utils import *
import os,sys,time,json
from os.path import exists
import copy
import pandas as pd
import more_itertools

print ("Logging in")
#Use dsl.ini file
dsl=dimcli.login()
dsl=dimcli.Dsl()
```

```python
#Top 1000 pubs by altmetric
highlight_pubs_altmetric=dsl.query(f'''search publications
  where year in [2022:2023]
  return publications [id+title+doi+journal+times_cited+altmetric+field_citation_ratio+relative_citation_ratio+funders]
  sort by altmetric
  limit 1000''').as_dataframe()
#  return funders''').as_dataframe()

highlight_pubs_altmetric
```

```python
#Export
highlight_pubs_altmetric.to_excel("OAC-Potential-Highlights.xlsx")
```

```python
#Top 1000 pubs by citations
highlight_pubs_citations=dsl.query(f'''search publications
  where year in [2022:2023]
  return publications [id+title+doi+journal+altmetric+times_cited+funders]
  sort by times_cited
  limit 1000''').as_dataframe()
#  return funders''').as_dataframe()

highlight_pubs_citations.dropna(inplace=True)
highlight_pubs_citations
#hpc.head()
```
